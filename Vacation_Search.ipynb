```python
# Import the dependencies
import pandas as pd
import gmaps
import requests

# Import the API key.
from config import g_key
```

```python
# Import weather database as a new dataframe
weather_py_df = pd.read_csv("weather_data/WeatherPy_database.csv")
weather_py_df.head()
```

```python
# Ask the customer to add a minimum and maximum temperature value
min_temp = float(input("What is the minimum temperature you would like for your trip? "))
max_temp = float(input("What is the maximum temperature you would like for your trip? "))
yes_rain = str(input("Do you want it to be raining?(yes/no)"))
yes_snow = str(input("Do you want it to be snowing?(yes/no)"))
```

```python
# Filter the dataset to find the cities that fit the criteria    
preferred_cities_temp_df = weather_py_df.loc[(weather_py_df["Max Temp"] <= max_temp) & \
                                       (weather_py_df["Max Temp"] >= min_temp)]

preferred_cities_temp_df.head(10)
```

```python
# Editing DataFrame for rain preference
if yes_rain == "no":
    preferred_cities_temp_df = preferred_cities_temp_df.loc[preferred_cities_temp_df["Rain(inches)"] == 0]
else:
    preferred_cities_temp_df = preferred_cities_temp_df.loc[preferred_cities_temp_df["Rain(inches)"] > 0]
preferred_cities_temp_df.head(10)
```

```python
# Editing DataFrame for snow preference
if yes_snow == "no":
    preferred_cities_temp_df = preferred_cities_temp_df.loc[preferred_cities_temp_df["Snow(inches)"] == 0]
else:
    preferred_cities_temp_df = preferred_cities_temp_df.loc[preferred_cities_temp_df["Snow(inches)"] > 0]
preferred_cities_temp_df.head(10)
```

```python
# Adding the cities to a marker layer map with a pop-up marker for each city that includes
hotel_df = preferred_cities_temp_df[["City", "Country", "Max Temp", "Current Description","Lat", "Lng"]].copy()
hotel_df["Hotel Name"] = ""
hotel_df.head(10)
```

```python
# Iterate through the DataFrame
for index, row in hotel_df.iterrows():
    
    # Get the latitude and longitude
    lat = row["Lat"]
    lng = row["Lng"]

    # Add the latitude and longitude to location key for the params dictionary
    params["location"] = f"{lat},{lng}"

    # Use the search term: "lodging" and our latitude and longitude
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    
    # Make request and get the JSON data from the search
    hotels = requests.get(base_url, params=params).json()
   
    # Grab the first hotel from the results and store the name.
    try:
        hotel_df.loc[index, "Hotel Name"] = hotels["results"][0]["name"]
    except (IndexError):
        print("Hotel not found... skipping.")
    
# Indicate that Data Loading is complete.
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```
